```python
import csv
import os
import subprocess
from string import punctuation

import pandas as pd

import spacy
import spacy_transformers
from wordfreq import zipf_frequency
```

```python
# 1. Redefine the function to explicitly handle international characters safely
def count_csv_elements_in_file(filepath):
    try:
        # 'utf-8' handles international languages.
        # errors='ignore' ensures it won't crash if it hits a stray unreadable byte
        with open (filepath, "r", encoding="utf-8", errors='ignore') as file:
              return len(file.read().splitlines())
    except Exception:
        return 0
```

```python
# 2. Re-run the data loop with defensive structure
language =[]
total_words =[]

for path, subdirs, files in os.walk("raw-word-lists"):
    for name in files:
        if name.endswith(".txt") or name.endswith(".csv"):
           filename = (os.path.join(path, name))

           lang_name = name.split(".")[0]
           count = count_csv_elements_in_file(filename)

           language.append(lang_name)
           total_words.append(count)

    print(f"Total Languages Successfully Read: {len(language)}")
    print(f"Total Word Counts Successfully Read: {len(total_words)}")
```

```python
 # Save the table into a variable called 'df'
df = pd.DataFrame({
    "language": language,
    "total_words": total_words
})

# Type 'df' on its own line to display it underneath the cell
df
```
